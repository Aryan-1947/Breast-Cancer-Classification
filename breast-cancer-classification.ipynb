```python
breast_cancer_dataset=pd.read_csv('/kaggle/input/breast-cancer-dataset/breast-cancer.csv')
breast_cancer_dataset.head()
```

```python
le=LabelEncoder()
breast_cancer_dataset['diagnosis']=le.fit_transform(breast_cancer_dataset['diagnosis'])
```

```python
breast_cancer_dataset.drop('id',axis=1,inplace=True)
```

```python
# Separating features and the target column
x=breast_cancer_dataset.iloc[:,1:]
y=breast_cancer_dataset['diagnosis']

print(x)
print(y)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
print(x.shape,x_train.shape,x_test.shape)
```

# **Model training**

```python
model=LogisticRegression()
```

```python
# Training the logistic regression model using the training data
model.fit(x_train,y_train)
```

```python
# accuracy on training data ( usually we dont do this we find accuracy and do prediction on test data)
x_pred=model.predict(x_train)
print(accuracy_score(y_train,x_pred))
```

```python
# accuracy on training data
y_pred=model.predict(x_test)
print(accuracy_score(y_test,y_pred))

# both accuracy are checked to chk overfitting
```
